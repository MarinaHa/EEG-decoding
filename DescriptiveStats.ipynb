## DescriptiveStats
##### Explores and visualizes features of the HCI data set. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Load dataframe
df = pd.read_csv('EmotionElicitation_vlncLevel')
df
```

```python
# Plot histograms for several parameters

for parameter in ['arslLevel', 'vlncLevel', 'subjectID', 'feltArsl', 'feltCtrl', 'feltEmo', 'feltPred', 'feltVlnc']: 
    plt.hist(df[parameter], bins=np.arange(min(df[parameter]), max(df[parameter]) + 2, 1))
    plt.xlabel(parameter)
    plt.ylabel('Frequency')
    fig = plt.gcf()
    plt.show()
```

```python

sns.set(style="white")

# Generate a large random dataset
d = df

# Compute the correlation matrix
corr = d.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
#f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

plt.show()
```
